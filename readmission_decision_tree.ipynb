## Splitting up the data
#### Creating a trainning and test data set

```python
# x = all columns BESIDES readmitted
# y = ONLY readmitted

X = df.drop(['readmitted'], axis = 1)
y = df.pop('readmitted')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
## Scaling data
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Decision Tree Analysis

A decision tree is a flowchart-like tree structure where an internal node represents feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node. It learns to partition on the basis of the attribute value. It partitions the tree in recursively manner call recursive partitioning. This flowchart-like structure helps you in decision making. It's visualization like a flowchart diagram which easily mimics the human level thinking. That is why decision trees are easy to understand and interpret.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = tree.DecisionTreeClassifier(max_depth = 1000)

# fit the model
clf.fit(X_train,y_train)

# model score
predict_train_clf = clf.predict(X_train)
predict_test_clf = clf.predict(X_test)

# accuracy score
clf_train_score = clf.score(X_train,y_train)
clf_test_score = clf.score(X_test,y_test)

# f1-score
clf_f1_score = metrics.f1_score(y_test, predict_test_clf, average = 'macro')

print('Accuracy on Train set',clf_train_score)
print('Accuracy on Test set',clf_test_score)
print('F1-score on Test set:',clf_f1_score)
print('\n')
print(metrics.classification_report(y_test, predict_test_clf))
```
